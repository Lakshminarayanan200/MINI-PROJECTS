```python
q1=df["Property_Sale_Price"].quantile(0.25)
q3=df["Property_Sale_Price"].quantile(0.75)
iqr=q3-q1
lower_lmt=q1-1.5*iqr
upper_lmt=q3+1.5*iqr
dataset=df[((df["Property_Sale_Price"]>lower_lmt) & (df["Property_Sale_Price"]))]

q1=dataset["LotFrontage"].quantile(0.25)
q3=dataset["LotFrontage"].quantile(0.75)
iqr=q3-q1
print(q1,q3)
lower_lmt=q1-1.5*iqr
upper_lmt=q3+1.5*iqr
dataset=df[((df["LotFrontage"]>lower_lmt) & (df["LotFrontage"]))]
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
def encode(data,columns):
  for column in columns:
    data[column]=le.fit_transform(data[column])
  return data
```

```python
columns=dataset.select_dtypes(include ='object')
encode(dataset,columns)
```

```python
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
print(x_train)
```

```python
print(y_train)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred))
```
